```python
import pandas as pd  # Make sure to import pandas as well
import spacy
from nltk.corpus import stopwords
import nltk

# Load English model
nlp = spacy.load('en_core_web_sm')

stop_words = set(stopwords.words('english'))  # Load stopwords

# Function to tokenize and lemmatize a single text, while removing stopwords
def tokenize_and_lemmatize_text(text):
    doc = nlp(text)  # Process the text with spaCy
    lemmatized_tokens = [token.lemma_.lower() for token in doc if token.is_alpha and token.lemma_.lower() not in stop_words]  # Lemmatize tokens and remove stopwords
    return ' '.join(lemmatized_tokens)

# Assuming 'new_df' is your DataFrame and 'All_text' is the column with text to process
# Apply the tokenization and lemmatization function to each row in the column
new_df['Processed'] = new_df['All_text'].apply(tokenize_and_lemmatize_text)
```
